```python
# 1. Загрузка датасета
from sklearn.datasets import load_iris
iris_dataset = load_iris()

# 2. Просмотр содержания датасета
print(iris_dataset.keys())

# 3. Просмотр классов в датасете (видов ириса)
print(iris_dataset['target_names'])

# 4. Просмотр признаков (features)
print(iris_dataset['feature_names'])

# 5. Создание датафрейма
import pandas as pd
iris_dataset_df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
print(iris_dataset_df.head(10))

# 6. Два датафрейма с признаками для удобства графического отображения данных
iris_dataset_1 = iris_dataset_df[['sepal length (cm)', 'petal length (cm)', 'petal width (cm)']]
iris_dataset_2 = iris_dataset_df[['sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]

# 7. Просмотр первых строк
print(iris_dataset_1.head(2))
print(iris_dataset_2.head(2))

# 8. Подключение библиотек для графиков
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

# 9. Построение 3D-графика для первого датафрейма
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = iris_dataset_1['sepal length (cm)']
xdata = iris_dataset_1['petal length (cm)']
ydata = iris_dataset_1['petal width (cm)']
ax.scatter3D(xdata, ydata, zdata, alpha=.8, c=iris_dataset.target)
plt.title('3D Scatter Plot: Dataset 1')
plt.show()

# 10. Построение 3D-графика для второго датафрейма
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = iris_dataset_2['sepal width (cm)']
xdata = iris_dataset_2['petal length (cm)']
ydata = iris_dataset_2['petal width (cm)']
ax.scatter3D(xdata, ydata, zdata, alpha=.8, c=iris_dataset.target)
plt.title('3D Scatter Plot: Dataset 2')
plt.show()

# 11. Загрузка k-NN и разделение датасета
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# 12. Разделение датасета на обучающую и тестовую части
x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(
    iris_dataset_1, iris_dataset.target, test_size=0.2, random_state=0)
x_train_2, x_test_2, y_train_2, y_test_2 = train_test_split(
    iris_dataset_2, iris_dataset.target, test_size=0.2, random_state=0)

# 13. Цикл для изменения числа соседей
from sklearn.metrics import accuracy_score

# Переменные для сохранения результата
neighbors_range = range(1, 21)
accuracies_1 = []
accuracies_2 = []

# Цикл по количеству соседей
for n_neighbors in neighbors_range:
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    
    # Тренировка на первом наборе признаков
    knn_model_1 = knn.fit(x_train_1, y_train_1)
    predictions_1 = knn_model_1.predict(x_test_1)
    accuracies_1.append(accuracy_score(y_test_1, predictions_1))
    
    # Тренировка на втором наборе признаков
    knn_model_2 = knn.fit(x_train_2, y_train_2)
    predictions_2 = knn_model_2.predict(x_test_2)
    accuracies_2.append(accuracy_score(y_test_2, predictions_2))

# Вывод результатов
print("Results for varying number of neighbors:")
for n, acc1, acc2 in zip(neighbors_range, accuracies_1, accuracies_2):
    print(f"{n} neighbors -> Accuracy_1: {acc1:.3f}, Accuracy_2: {acc2:.3f}")

# Визуализация зависимости качества от числа соседей
plt.plot(neighbors_range, accuracies_1, label='Dataset 1')
plt.plot(neighbors_range, accuracies_2, label='Dataset 2')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy vs Number of Neighbors')
plt.show()
```

```python
# 1. Загрузка датасета
from sklearn.datasets import load_iris
iris_dataset = load_iris()

# 2. Просмотр содержания датасета
print(iris_dataset.keys())

# 3. Просмотр классов в датасете (видов ириса)
print(iris_dataset['target_names'])

# 4. Просмотр признаков (features)
print(iris_dataset['feature_names'])

# 5. Создание датафрейма
import pandas as pd
iris_dataset_df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset['feature_names'])
print(iris_dataset_df.head(10))

# 6. Два датафрейма с признаками для удобства графического отображения данных
iris_dataset_1 = iris_dataset_df[['sepal length (cm)', 'petal length (cm)', 'petal width (cm)']]
iris_dataset_2 = iris_dataset_df[['sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]

# 7. Просмотр первых строк
print(iris_dataset_1.head(2))
print(iris_dataset_2.head(2))

# 8. Подключение библиотек для графиков
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

# 9. Построение 3D-графика для первого датафрейма
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = iris_dataset_1['sepal length (cm)']
xdata = iris_dataset_1['petal length (cm)']
ydata = iris_dataset_1['petal width (cm)']
ax.scatter3D(xdata, ydata, zdata, alpha=.8, c=iris_dataset.target)
plt.title('3D Scatter Plot: Dataset 1')
plt.show()

# 10. Построение 3D-графика для второго датафрейма
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = iris_dataset_2['sepal width (cm)']
xdata = iris_dataset_2['petal length (cm)']
ydata = iris_dataset_2['petal width (cm)']
ax.scatter3D(xdata, ydata, zdata, alpha=.8, c=iris_dataset.target)
plt.title('3D Scatter Plot: Dataset 2')
plt.show()

# 11. Реализация собственного метода k-NN
from scipy.stats import mode

def my_knn_predict(X_train, y_train, X_test, k=5):
    """
    Реализация алгоритма k-ближайших соседей для классификации.
    
    Аргументы:
    X_train - обучающие данные (массив размера NxM, где N - количество обучающих примеров, M - количество признаков)
    y_train - метки классов обучающих данных (массив длины N)
    X_test - тестовые данные, для которых нужно предсказать классы (массив размера TxM)
    k - количество ближайших соседей, которые будут использоваться для классификации
    
    Возвращает:
    predictions - массив предсказанных классов для тестовых данных
    """
    
    # Список для хранения предсказанных классов для всех точек тестовой выборки
    predictions = []
    
    # Проходим по каждой точке тестовой выборки
    for test_point in X_test:
        # 1. Вычисляем расстояния между текущей тестовой точкой и всеми точками обучающей выборки.
        # Используется евклидово расстояние (норма L2).
        distances = np.linalg.norm(X_train - test_point, axis=1)
        
        # 2. Находим индексы k ближайших соседей (точек с наименьшими расстояниями).
        k_indices = np.argsort(distances)[:k]
        
        # 3. Получаем метки классов для этих k ближайших соседей.
        k_labels = y_train[k_indices]
        
        # 4. Определяем наиболее часто встречающийся класс среди соседей (моду).
        # Используем scipy.stats.mode с keepdims=True, чтобы вернуть корректный массив.
        pred_class = mode(k_labels, keepdims=True).mode[0]
        
        # 5. Добавляем предсказанный класс в список предсказаний.
        predictions.append(pred_class)
    
    # Преобразуем список предсказаний в numpy-массив для удобства дальнейшей обработки.
    return np.array(predictions)


# 12. Разделение датасета на обучающую и тестовую части
from sklearn.model_selection import train_test_split

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(
    iris_dataset_1.values, iris_dataset.target, test_size=0.2, random_state=0)
x_train_2, x_test_2, y_train_2, y_test_2 = train_test_split(
    iris_dataset_2.values, iris_dataset.target, test_size=0.2, random_state=0)

# 13. Цикл для изменения числа соседей с использованием собственного метода
from sklearn.metrics import accuracy_score

# Переменные для сохранения результата
neighbors_range = range(1, 21)
accuracies_1 = []
accuracies_2 = []

# Цикл по количеству соседей
for k in neighbors_range:
    # Предсказания для первого набора признаков
    predictions_1 = my_knn_predict(x_train_1, y_train_1, x_test_1, k=k)
    accuracies_1.append(accuracy_score(y_test_1, predictions_1))
    
    # Предсказания для второго набора признаков
    predictions_2 = my_knn_predict(x_train_2, y_train_2, x_test_2, k=k)
    accuracies_2.append(accuracy_score(y_test_2, predictions_2))

# Вывод результатов
print("Results for varying number of neighbors (custom k-NN):")
for k, acc1, acc2 in zip(neighbors_range, accuracies_1, accuracies_2):
    print(f"{k} neighbors -> Accuracy_1: {acc1:.3f}, Accuracy_2: {acc2:.3f}")

# Визуализация зависимости качества от числа соседей
plt.plot(neighbors_range, accuracies_1, label='Dataset 1 (Custom k-NN)')
plt.plot(neighbors_range, accuracies_2, label='Dataset 2 (Custom k-NN)')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy vs Number of Neighbors (Custom k-NN)')
plt.show()
```
